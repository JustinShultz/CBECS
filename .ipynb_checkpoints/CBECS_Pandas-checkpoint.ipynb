# Commercial Buildings Energy Consumption Survey (CBECS) Pandas Data Analyses

```python
import pandas as pd 
import requests 
from os.path import exists
import matplotlib.pyplot as plt 
# import chart_studio.plotly as py
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot
import plotly.express as px

init_notebook_mode(connected=True)         # initiate notebook for offline plot

%config Completer.use_jedi = False         # turn of jedi so that autocomplete is faster

pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', 255)
# pd.set_option('display.auto_truncate_string_exceeding_this_length', 255)
```

# Microdata

```python
if exists('cbecs2012_microdata.csv') == False:
    download_url = 'https://www.eia.gov/consumption/commercial/data/2012/xls/2012_public_use_data_aug2016.csv'
    
    cbecs_response = requests.get(download_url)

    with open('cbecs2012_microdata.csv', 'wb') as f:
        f.write(cbecs_response.content)
    
cbecs = pd.read_csv('cbecs2012_microdata.csv')
cbecs.head()
```

```python
cbecs.info()
```

```python
cbecs.shape
```

# Variables

```python
if exists('cbecs2012_variables.xlsx') == False:
    variables_url = 'https://www.eia.gov/consumption/commercial/data/2012/xls/2012microdata_codebook.xlsx'

    variable_response = requests.get(variables_url)

    with open('cbecs2012_variables.xlsx', 'wb') as f:
        f.write(variable_response.content)
    
variables = pd.read_excel('cbecs2012_variables.xlsx')
variables.head(10)
```

```python
# variables.loc[variables['Label'].str.contains('type', case=False)]
variables.loc[variables['Label'].str.contains('consumption', case=False)]
```

```python
variables.loc[variables['Label'].str.contains('foot', case=False)]
```

```python
# print(variables.loc[variables['Variable\nname'] == 'PBA']['Values/Format codes'].values
# variables.loc[variables['Variable\nname'] == 'PBA']['Values/Format codes'].str.split('\n')
activity_dict = {1: 'Vacant', 2: 'Office', 4: 'Laboratory', 5: 'Nonrefrigerated warehouse', 6: 'Food sales', 7: 'Public order and safety', 8: 'Outpatient health care', 11: 'Refrigerated warehouse', 12: 'Religious worship', 13: 'Public assembly', 14: 'Education', 15: 'Food service', 16: 'Inpatient health care', 17: 'Nursing', 18: 'Lodging', 23: 'Strip shopping mall', 24: 'Enclosed mall', 25: 'Retail other than mall', 26: 'Service', 91: 'Other'} 
cbecs['BTYPE'] = cbecs['PBA']
cbecs['BTYPE'] = cbecs['BTYPE'].map(activity_dict).fillna(cbecs['BTYPE'])
```

```python
variables.loc[variables['Label'].str.contains('foot', case=False)]['Values/Format codes'].values
```

# Add Energy Use Intensity (EUI)

```python
cbecs['EUI_KBTU'] = cbecs['MFBTU'] / cbecs['SQFT']
cbecs.info()
```

# Boxplot

```python
# cbecs.boxplot(by='BTYPE', column=['EUI_KBTU'], figsize=(24,12), rot=45)
boxplot_fig = px.box(cbecs, x='BTYPE', y="EUI_KBTU")
boxplot_fig.show()
```

# Offices

```python
offices = cbecs.loc[cbecs['PBA'] == 2].copy()
offices.shape
```

```python
offices['PUBID_str'] = offices['PUBID'].astype(str)
offices['PUBID_str'].dtypes
```

```python
# offices
# offices.reindex(offices['PUBID_str'])
```

```python
# offices.sort_values(by=['SQFT']).plot.bar(x='PUBID_str', y='MFBTU', rot=0, figsize=(24,12))
offices_bySQFT = offices.sort_values(by=['SQFT']).copy()
# offices_bySQFT = offices.sort_values(by=['SQFT']).loc[offices['SQFT'] > 100000].copy()
offices_bars = px.scatter(offices_bySQFT, x='SQFT', y='EUI_KBTU', color='SQFTC', labels={'SQFTC': 'Category'}, trendline='ols')
offices_bars.show()
```

```python
offices_bySQFT.groupby('SQFTC')['EUI_KBTU'].describe()
```

```python
offices.sort_values(by=['SQFT']).tail(25).plot.bar(x='PUBID_str', y='MFBTU', rot=0, figsize=(24,12))
```
